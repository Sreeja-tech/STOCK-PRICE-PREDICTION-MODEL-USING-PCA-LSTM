```python
#Importing all the required libraries
import numpy as np
import pandas as pd
from numpy import array
from numpy import mean
from numpy import cov
from numpy.linalg import eig
```

```python
#Reading and Storing the dataset in variable named HDFC_df
HDFC_df = pd.read_csv("HDB.csv")
```

```python
#Printing to have a quick look
HDFC_df
```

```python
#Dropping the NA values to ensure accurate results
HDFC_df = HDFC_df.dropna()
```

```python
#Again printing
HDFC_df
```

```python
#Standardising the data to ensure it lies on base axis (0,0)
from sklearn.preprocessing import StandardScaler
std_obj = StandardScaler()
```

```python
#Using iloc to select only the required columns from the dataset
#We have dropped the date column
#Storing the standardised values in the same variable again
HDFC_df.iloc[:,1:7] = std_obj.fit_transform(HDFC_df.iloc[:, 1:7])
HDFC_df.head()
```

```python
#Assigning variable x to the selected columns
#The colon at the start of the [] reprsents that we are selcting all the rows
x = HDFC_df.iloc[:, 1:7]
x
```

```python
#Importing sklearn library
import sklearn
```

```python
#Choosing PCA from sklearn
from sklearn.decomposition import PCA
```

```python
#Assigning PCA to pca
#The (0.9) represents that we are only capturing the features giving us 90% variability
#Fitting the data in x to our pca model
pca = PCA(0.9)
pca.fit(x)
```

```python
#Printing the eigen vectors transposed 
print(pca.components_.T)
```

```python
#Printing the eigen vectors
print(pca.components_)
```

```python
#Printing the eigen values
print(pca.explained_variance_)
```

```python
#Printing the percent variabilty captured by each eigen value
#This is given by the formula = (eigen value/sum of eigen values)
print(pca.explained_variance_ratio_)
```

```python
#This is the Z-Score or PCA1, PCA2 values
z = pca.transform(x)
print(z)
```
